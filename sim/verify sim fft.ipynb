```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.concatenate((1023*np.ones(8, dtype='int16'), 
                    -1022*np.ones(8, dtype='int16')))
x = np.concatenate((x, x))
```

```python
plt.plot(x, 'x-')
plt.title("input")
```

```python
def plot_complex(X, ax, title):
    re = ax.plot(np.real(X), 'x-', label='real')
    im = ax.plot(np.imag(X), 'x-', label='imag')
    ax.legend()
    ax.set_title(title)
    return re, im
```

## numpy output

```python
X = np.fft.fft(x)
```

```python
fig, ax = plt.subplots()
plot_complex(X, ax, "expected out")
```

## sim output

```python
# https://stackoverflow.com/questions/6727875/hex-string-to-signed-int-in-python-3-2
def to_int(hexstr, bits):
    value = int(hexstr, 16)
    if value & (1 << (bits - 1)):
        value -= 1 << bits
    return value
    
    
out = np.loadtxt("../fpga/simulation/modelsim/rom/fft_test_out.memh",
                 dtype='complex',
                 converters={0:lambda s: to_int(s[0:4], 16) + 1j*to_int(s[4:8], 16)})
```

```python
fig, ax = plt.subplots()
plot_complex(out, ax, "real out")
```

## Real output

```python
out = np.loadtxt("out.memh",
                 dtype='complex',
                 converters={0:lambda s: to_int(s[0:4], 16) + 1j*to_int(s[4:8], 16)})

fig, ax = plt.subplots()
plot_complex(out[0:15], ax, "output")
```

```python
np.argmax(np.abs(out[0:15]))
```
